```python
def gene_summary(fa, fb):
    mfa = ClusteringReader.read(fa)
    mfb = ClusteringReader.read(fb)
    try:
        hfa = HarmonizedClusteringResult(mfa)
    except:
        print('Unable to harmonize', fa)
        hfa = mfa
    try:
        hfb = HarmonizedClusteringResult(mfb)
    except:
        print('Unable to harmonize', fb)
        hfb = mfb

    fig = ComparativePlot(hfa, hfb, 'best')
    fig.save(path.expanduser(path.join('~/Desktop/Genes', g + '_compare.pdf')))
    fig.close()

def gene_samples(f):
    mf = ClusteringReader.read(f)
    return  mf.cells.index.unique('Sample')
```

```python
sample_dir = path.expanduser(
    '~/Google Drive File Stream/My Drive/Projects/RDN-WDP/clustering-h5')
files = os.listdir(sample_dir)
sample_files = []
for sample_f in files:
    if path.isfile(path.join(sample_dir, sample_f)) and sample_f.endswith('.h5') and not sample_f.endswith('f.h5'):
        sample_files.append(str(path.join(sample_dir, sample_f)))

for sample_file in sample_files:
    g = os.path.splitext(os.path.basename(sample_file))[0]
    gf = g + 'f'
    print("Processing a=", str(g) + ", b=" + str(gf) + "...")
    try:
        gene_summary(str(path.join(sample_dir, g)) + '.h5', str(path.join(sample_dir, gf)) + '.h5')
    except:
        continue
```
